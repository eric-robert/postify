## Style

```python
Pipeline.InputFiles('../workspace/animals/cropped/lion.*')
```

```python
def aprox ():
    Scale.Up(width=2000, height=2000)
    Procedure.AproxBox(background=Color.black, box_density=0.2, high_detail=True, callback=Effects.blur)

Pipeline.RunWith(aprox)
Pipeline.Show()
```

```python
Pipeline.SaveFiles('../workspace/animals/styled')
```

## To Posters

```python
Pipeline.InputFiles('../workspace/animals/styled/lion.*')

def poster():
    Effects.denoise()
    Effects.denoise()
    Effects.denoise()
    Border.Even(color=Color.black, size=0.2, resize=True)
    Border.Blurred(size=0.17, intensity=0.02)
    Border.Blurred(size=0.18, intensity=0.01)
    Border.Blurred(size=0.19, intensity=0.005)
    Border.Blurred(size=0.20, intensity=0.002)

    To.notebook(size=15)

Pipeline.RunWith(poster)
```
